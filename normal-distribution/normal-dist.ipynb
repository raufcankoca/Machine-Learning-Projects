## Normal Distribution

Normal distribution, also known as the Gaussian distribution or bell curve, is a statistical probability distribution that is widely used in various fields, including statistics, mathematics, and natural sciences. It is a continuous probability distribution characterized by its symmetric, bell-shaped curve.

The key features of a normal distribution are as follows:

1. Symmetry: The curve of a normal distribution is symmetric around its mean. This means that the mean, median, and mode of the distribution are all equal and located at the center of the curve.

2. Bell-shaped curve: The shape of the normal distribution resembles a bell, with the highest point at the mean and the curve gradually tapering off on either side. The curve is smooth and continuous, extending indefinitely in both directions.

3. Mean and standard deviation: The distribution is defined by two parameters: the mean () and the standard deviation (). The mean determines the central location of the curve, while the standard deviation controls the spread or dispersion of the data around the mean.

4. Empirical Rule: In a normal distribution, approximately 68% of the data falls within one standard deviation of the mean, about 95% falls within two standard deviations, and approximately 99.7% falls within three standard deviations. This is known as the empirical rule or the 68-95-99.7 rule.




The probability density function (PDF) of the normal distribution is given by the equation:

$$f(x) = \frac{1}{\sigma \sqrt{2\pi}} e^{-\frac{(x - \mu)^2}{2\sigma^2}}$$


where f(x) represents the probability density at a given value x,  is the mean,  is the standard deviation, and e is the base of the natural logarithm.

The normal distribution has several important properties that make it useful in statistical analysis. It is often used as a theoretical model for many real-world phenomena, as it approximates the distribution of many naturally occurring variables, such as heights, weights, test scores, and measurement errors. Additionally, it plays a fundamental role in statistical inference and hypothesis testing, where assumptions of normality are often made.

In summary, the normal distribution is a symmetric, bell-shaped probability distribution characterized by its mean and standard deviation. It is widely used to describe and analyze data in various fields due to its mathematical properties and its close approximation to many real-world phenomena.

# PROBLEM : The exam results in the Electromagnetics course are normally distributed. The average grade of the students is 45 with a standard deviation of 20. What is the probability of a student scoring above 80?

```python
import scipy.stats as stats

mean = 45
std_dev = 20
x = 80

# Z-skorunu hesapla
z_score = (x - mean) / std_dev

# Z-skoruna karlk gelen olasl bul
probability = 1 - stats.norm.cdf(z_score)

print("The probability of a student getting more than 80 grades:", probability)
```

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

mean = 45
std_dev = 20
x = 80

# Olaslk younluk fonksiyonunu (PDF) ve kmlatif dalm fonksiyonunu (CDF) hesapla
x_values = np.linspace(mean - 4 * std_dev, mean + 4 * std_dev, 100)
pdf = stats.norm.pdf(x_values, mean, std_dev)
cdf = stats.norm.cdf(x_values, mean, std_dev)

# X=80 noktasnn zerini boyayarak olasl gster
plt.figure(figsize=(10, 5))
plt.plot(x_values, pdf, label='PDF')
plt.fill_between(x_values, pdf, where=(x_values >= x), alpha=0.5, color='orange')
plt.title('Normal Distribution')
plt.xlabel('Grades')
plt.ylabel('Probability Density Function (PDF)')
plt.legend()
plt.grid(True)
plt.show()

# Kumulatif dalm fonksiyonunu (CDF) gster
plt.figure(figsize=(10, 5))
plt.plot(x_values, cdf, label='CDF')
plt.axvline(x, color='r', linestyle='--', label='X = 80')
plt.title('Normal Distribution')
plt.xlabel('Grades')
plt.ylabel('Cumulative Density Function (PDF)')
plt.legend()
plt.grid(True)
plt.show()
```

# Lognormal distribution

The lognormal distribution is a probability distribution that is derived from the logarithm of a random variable that follows a normal distribution. In other words, if a random variable X follows a normal distribution, then its logarithm, Y = ln(X), follows a lognormal distribution.

The lognormal distribution is characterized by the following properties:

1. Positivity: The lognormal distribution is defined for positive values only. Since it is derived from the logarithm of a random variable, the values of the lognormal distribution are always greater than zero.

2. Skewness: The lognormal distribution is typically positively skewed. This means that the tail on the right side of the distribution is longer than the left side. The skewness arises from the fact that the logarithm of a positive value will always result in a negative or zero value, and therefore, the distribution is stretched out towards larger values.

3. Multiplicative nature: The lognormal distribution is often used to model variables that are the product of several independent random factors. For example, if a variable is the result of multiplying multiple random variables together, such as the growth rate of a stock over several time periods, it is often more appropriate to model it with a lognormal distribution.



The probability density function (PDF) of the lognormal distribution is given by the equation:

$$f(x) = \frac{1}{{x \cdot \sigma \sqrt{2\pi}}} \cdot e^{-\frac{(\ln(x) - \mu)^2}{2\sigma^2}}$$

where f(x) represents the probability density at a given value x,  is the mean of the logarithm of the random variable,  is the standard deviation of the logarithm of the random variable, ln(x) is the natural logarithm of x, and e is the base of the natural logarithm.

The lognormal distribution has applications in various fields, including finance, economics, and biology. It is commonly used to model variables such as stock prices, asset returns, income, population sizes, and the size of particles in aerosols.

It is worth noting that while the normal distribution is symmetric, the lognormal distribution is not. The lognormal distribution describes variables that tend to have a wide range of positive values and can have a significant impact on statistical analysis and modeling when dealing with skewed data.

## Why do we need to convert lognormal dist. to normal dist. ?

Converting a dataset with a lognormal distribution to a normal distribution can be beneficial in certain statistical analyses and modeling scenarios. Here are a few reasons why such a conversion may be desirable:

1. Simplicity of analysis: The normal distribution is well-studied and has many convenient mathematical properties. Many statistical methods and techniques, such as hypothesis testing and regression analysis, are based on the assumption of normality. By transforming a lognormally distributed dataset into a normal distribution, you can apply these standard statistical techniques more easily.

2. Validating assumptions: Certain statistical models, such as linear regression, assume that the error terms or residuals follow a normal distribution. If the response variable or any predictors in the model have a lognormal distribution, converting them to a normal distribution helps meet the assumptions of the model and ensures the validity of the results.

3. Data interpretation and visualization: Lognormally distributed data can be challenging to interpret and visualize due to the asymmetry of the distribution. Transforming the data to a normal distribution can make the patterns and relationships in the data more apparent and facilitate easier interpretation and visualization.

4. Overcoming heteroscedasticity: In some cases, lognormality may be a result of heteroscedasticity, where the variability of the data increases or decreases as the values change. Transforming the data to a normal distribution can help stabilize the variance and address the issue of heteroscedasticity, making it easier to apply various statistical techniques that assume constant variance.

It's important to note that the decision to convert a lognormal distribution to a normal distribution should be made based on the specific requirements and assumptions of the analysis or modeling task at hand. It is crucial to consider the underlying characteristics of the data and the implications of the transformation on the results.
